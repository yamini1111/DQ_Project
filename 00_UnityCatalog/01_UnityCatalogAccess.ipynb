```python
catalog_name = "ucdqdev"
```

```python
spark.sql(f"""GRANT USAGE, BROWSE
ON CATALOG {catalog_name}
TO `account users`""")
```

```python
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS {catalog_name}.bronze""")
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS {catalog_name}.silver""")
```

```python
spark.sql(f"""GRANT USAGE, CREATE, MODIFY
ON SCHEMA {catalog_name}.bronze
TO `account users`""")
```

```python
spark.sql(f"""GRANT CREATE
ON SCHEMA {catalog_name}.bronze
TO `account users`""")
```

```python
spark.sql(f"""
GRANT ALL PRIVILEGES
ON SCHEMA {catalog_name}.bronze
TO `account users`""")
```

```python
spark.sql(f"""
GRANT ALL PRIVILEGES
ON SCHEMA {catalog_name}.silver
TO `account users`""")
```

```python
spark.sql("""GRANT READ FILES
ON EXTERNAL LOCATION raw
TO `account users`""")
```

```python
spark.sql("""GRANT WRITE FILES
ON EXTERNAL LOCATION raw
TO `account users`""")
```

```python
spark.sql("""
GRANT CREATE EXTERNAL TABLE 
ON EXTERNAL LOCATION raw
TO `account users`""")
```

```python
spark.sql(f"""CREATE SCHEMA IF NOT EXISTS {catalog_name}.dataquality""")
spark.sql(f"""CREATE VOLUME IF NOT EXISTS {catalog_name}.dataquality.dqcheckbadresults""")
spark.sql(f"""CREATE VOLUME IF NOT EXISTS {catalog_name}.dataquality.dqcheckbadrecords""")
```

```python
spark.sql(f"""GRANT READ VOLUME ON VOLUME {catalog_name}.bronze.voldqlibraries TO `account users`""")
```
